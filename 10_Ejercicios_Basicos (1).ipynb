```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-white")
import numpy as np
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
# Generar datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Crear figura y subplots
plt.figure(figsize=(10, 5))  # Tamao de la figura

# Primer subplot: Seno
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer grfico
plt.plot(x, y_sin)
plt.title("Seno")

# Segundo subplot: Coseno
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo grfico
plt.plot(x, y_cos)
plt.title("Coseno")

# Mostrar la figura
plt.tight_layout()  # Ajustar el espacio entre los subplots
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
# Generar datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# Crear figura y ejes
fig, axes = plt.subplots(3, 1, figsize=(8, 8))  # 3 filas, 1 columna

# Subplot 1: Seno
axes[0].plot(x, y_sin, color="red")
axes[0].set_title("Seno")
axes[0].set_ylim(-1.5, 1.5)  # Limitar el eje Y

# Subplot 2: Coseno
axes[1].plot(x, y_cos, color="green")
axes[1].set_title("Coseno")
axes[1].set_ylim(-1.5, 1.5)  # Limitar el eje Y

# Subplot 3: Tangente
axes[2].plot(x, y_tan, color="blue")
axes[2].set_title("Tangente")
axes[2].set_ylim(-40, 40)  # Limitar el eje Y para evitar distorsiones

# Ajustar diseo
plt.tight_layout()
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Lnea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
# Generar datos
x = np.linspace(0, 10, 100)
y_sin = np.sin(3 * x)  # Escalar la frecuencia
y_cos = np.cos(3 * x)  # Escalar la frecuencia

# Crear figura y subplots
plt.figure(figsize=(10, 5))  # Tamao de la figura

# Primer subplot: Seno
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer grfico
plt.plot(x, y_sin, linestyle='--', color='magenta')  # Lnea discontinua y color magenta
plt.ylim(-1, 1)  # Limitar el eje Y

# Segundo subplot: Coseno
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo grfico
plt.plot(x, y_cos, linestyle='-.', color='cyan')  # Lnea discontinua-punteada y color cyan
plt.ylim(-1, 1)  # Limitar el eje Y

# Ajustar el diseo
plt.tight_layout()
plt.show()
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperblico y coseno hiperblico se obtienen as:
* np.sinh, seno hiperblico;
* np.cosh, coseno hiperblico;


```python
# Generar datos
x = np.linspace(0, 10, 100)
y_sinh = np.sinh(x)  # Seno hiperblico
y_cosh = np.cosh(x)  # Coseno hiperblico

# Crear figura y subplots
fig, axes = plt.subplots(2, 1, figsize=(8, 6))  # 2 filas, 1 columna

# Primer subplot: Seno hiperblico
axes[0].plot(x, y_sinh, color='red')  # Lnea roja
axes[0].set_title("Seno Hiperblico")  # Ttulo
axes[0].set_xlabel("x")  # Etiqueta del eje X
axes[0].set_ylabel("sinh(x)")  # Etiqueta del eje Y

# Segundo subplot: Coseno hiperblico
axes[1].plot(x, y_cosh, color='blue')  # Lnea azul
axes[1].set_title("Coseno Hiperblico")  # Ttulo
axes[1].set_xlabel("x")  # Etiqueta del eje X
axes[1].set_ylabel("cosh(x)")  # Etiqueta del eje Y

# Ajustar diseo
plt.tight_layout()
plt.show()
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python
# Generar datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Crear figura y subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # 2 filas, 2 columnas

# Subplot 1: Seno
axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title("Seno")

# Subplot 2: Coseno
axes[0, 1].plot(x, y2, color='green')
axes[0, 1].set_title("Coseno")

# Subplot 3: Tangente
axes[1, 0].plot(x, y3, color='red')
axes[1, 0].set_title("Tangente")
axes[1, 0].set_ylim(-10, 10)  # Limitar valores para evitar distorsin

# Subplot 4: Exponencial
axes[1, 1].plot(x, y4, color='purple')
axes[1, 1].set_title("Exponencial")

# Ajustar espaciado
plt.subplots_adjust(hspace=0.4, wspace=0.3)  # Ajustar espaciado entre subplots

# Mostrar figura
plt.show()
```


### Ejercicio 6: 

Reproduce la figura siguiente empleando slo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros grficos (fila 1). Luego usa el estilo con mtodos (objetos `axes`) para aadir el tercer grfico.

```python
# Generar datos
x = np.linspace(0, 10, 100)
y_sqrt = np.sqrt(x)  # Raz cuadrada
y_cos = np.cos(x)    # Coseno
y_sin = np.sin(x)    # Seno

# Crear figura para los dos primeros grficos (Estilo MATLAB)
plt.figure(figsize=(8, 8))

# Grfico 1: Coseno y raz cuadrada
plt.subplot(2, 1, 1)  # 2 filas, 1 columna, posicin 1
plt.plot(x, y_cos, label="coseno", color="blue")
plt.plot(x, y_sqrt, label="raiz cuadrada", color="orange")
plt.title("Raz Cuadrada y coseno")
plt.legend()

# Crear el tercer grfico con el estilo basado en objetos (Axes)
fig, ax = plt.subplots(1, 1, figsize=(8, 3))  # Una fila, una columna
ax.plot(x, y_sin, color="blue")  # Dibujar el grfico del seno
ax.set_title("Seno")
ax.set_xlabel("x")
ax.set_ylabel("sen(x)")

# Ajustar diseo
plt.tight_layout()
plt.show()
```
